**10.  What are the key assumptions of Multiple Linear Regression?**

   -  Multiple linear regression requires that the relationship between predictors and the outcome is linear, errors are normally distributed with a mean of zero, and variance is constant (homoscedasticity). Key requirements include independence of observations, no perfect multicollinearity among predictors, and no significant endogeneity.

**11.   What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?**

  -  Heteroscedasticity is a condition in regression analysis where the variance of the error terms (residuals) is not constant across all values of the independent variables.This concept is important in Statistics and Machine Learning, especially when working with Multiple Linear Regression models.

     -Effect on Multiple Linear Regression:

     Heteroscedasticity does not bias the regression coefficients, but it creates several problems:

     1 Incorrect standard errors


     2 Unreliable hypothesis tests (t-test and F-test)


     3 Confidence intervals become inaccurate


     4 Model predictions become less reliable

**14.   What is the role of interaction terms in Multiple Linear Regression?**

  -  Interaction terms in multiple linear regression model the combined effect of two or more independent variables on a dependent variable, testing if the relationship between one predictor and the target changes based on the value of another variable. They allow for non-parallel regression slopes, improving model flexibility and accuracy.

**15.  How can the interpretation of intercept differ between Simple and Multiple Linear Regression?**

  -  Simple Linear Regression (SLR): The intercept represents the baseline value of Y when the single predictor X is zero. If X cannot be zero (e.g., weight), the intercept often lacks practical, independent meaning, acting only as an anchor for the line.

     Multiple Linear Regression (MLR): The intercept represents the expected value of Y when all included predictor variables are simultaneously zero. This intercept is heavily influenced by the presence of other variables in the model.

**16.  What is the significance of the slope in regression analysis, and how does it affect predictions?**

A steeper slope indicates a faster rate of change, while a positive/negative sign indicates direction. It affects predictions by determining the magnitude and direction of forecasted y-values based on input x-values.

**19.  How would you interpret a large standard error for a regression coefficient?**

  -  A large standard error for a regression coefficient indicates low precision and high uncertainty in the estimate, meaning the coefficient might vary significantly across different samples. It suggests that the predictor variable has a weak relationship with the response or, in multiple regression, indicates potential multicollinearity, making the estimate statistically insignificant.

**20.  How can heteroscedasticity be identified in residual plots, and why is it important to address it?**

  -  Heteroscedasticity is identified in residual plots by a distinct fan or cone-shaped pattern, where the vertical spread of residuals (error terms) systematically increases or decreases as predicted values (fitted values) rise. Addressing it is crucial because it violates OLS assumptions, leading to inefficient parameter estimates, invalid t-tests, and misleading confidence intervals.

**21.  What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?**

  -  AI Overview
A high R² paired with a low adjusted R² in multiple linear regression typically indicates that the model is overfitted, having too many predictors relative to the number of observations. The high R² suggests good fit on training data, while the low adjusted R² reveals that several predictors are irrelevant, adding noise rather than meaningful explanatory power.

**22.  Why is it important to scale variables in Multiple Linear Regression?**

 -  Scaling variables in multiple linear regression is important to ensure all features contribute equally to the model, speed up gradient descent convergence, and make coefficients directly comparable for identifying feature importance.

**23.  What is polynomial regression?**

  -  Polynomial regression is a form of regression analysis that models non-linear relationships between independent and dependent variables by fitting a curved line (nth-degree polynomial) to the data.

**24.  How does polynomial regression differ from linear regression?**

     -Relationship between variables is linear.

**25.  When is polynomial regression used?**

  -  Polynomial regression is used when data shows a non-linear relationship (e.g., curves, parabolas) that simple linear regression cannot accurately capture. It is best applied to model complex trends, such as increasing growth rates, or when residuals from a linear model display a pattern.

**27.  Can polynomial regression be applied to multiple variables?**

  -  Yes, polynomial regression can be applied to multiple variables, a technique known as multivariate polynomial regression. It models complex, non-linear relationships by creating polynomial features (squares, cubes, interaction terms like x1,x2)from multiple independent variables, which are then treated as distinct variables in a standard linear regression model.

**28.  What are the limitations of polynomial regression?**

  -  Polynomial regression is primarily limited by high risk of overfitting with high-degree polynomials, poor extrapolation performance outside the data range, and increased sensitivity to outliers.

**29.  What methods can be used to evaluate model fit when selecting the degree of a polynomial?**

  -  Selecting the optimal polynomial degree involves balancing model complexity with fit quality to avoid underfitting or overfitting. Key methods include k-fold cross-validation to assess performance on unseen data, analyzing residual plots for random patterns, and using information criteria like AIC/BIC. Often, a low-degree polynomial (degree 2 or 3) is a good starting point.

**30.  Why is visualization important in polynomial regression?**

  -  Visualization is critical in polynomial regression to detect non-linear patterns, verify the appropriate polynomial degree, and prevent overfitting. It helps identify outliers, assess how well the curved model fits the data, and ensures the model generalizes well rather than just passing through training points
